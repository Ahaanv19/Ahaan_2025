## Key Takeaways

1. **2D Array Initialization**: Creating a 2D boolean array with `new boolean[numRows][numCols]` and filling it with nested for loops is the standard approach. Every cell needs to be visited individually.

2. **Math.random() for Probability**: To give each cell a 40% chance of being true, check `Math.random() < 0.4`. Each call generates a fresh independent random value, so every cell gets its own roll.

3. **Traversing a Single Column**: To count values in one column of a 2D array, loop through rows while keeping the column index fixed: `lights[r][col]`. It's easy to accidentally loop over columns instead.

4. **Modulus for Divisibility Checks**: Use `%` (not `/`) to check if a number is even (`% 2 == 0`) or divisible by three (`% 3 == 0`). This came up in both rules of `evaluateLight`.

5. **Multi-Rule Return Logic**: When a method has multiple conditional rules with a default fallback, handle each specific case with its own if-statement and return the default at the end. Trying to combine everything into one expression usually leads to mistakes.

## Area of Stuggle 

My main struggle was mixing up rows and columns, both when creating the array in Part A and when traversing a column in Part B. I also found it cleaner to handle each rule with separate if-statements rather than one big expression.
